### yfinance 사용

```python
import yfinance as yf

# Microsoft (MSFT)에 대한 Ticker 객체 생성
msft = yf.Ticker("MSFT")

# Ticker 객체에 대한 정보 출력 (.py에서 실행할 때는 print(msft.info)로 사용)
display(msft.info)
```

```python
hist = msft.history(period="5d") # 5일간의 주가 데이터를 가져옴, 2개월:2mo, 1년:1y
display(hist) # 데이터 출력
```

```python
# recommandation :주식 분석가들이 이 주식에 대해 내린 추천（예 : 매수, 매도, 보유 등）을 포함하는 데이터프레임 반환
# period는 분석가들이 추천한 시점

msft.recommendations # 추천 정보 출력, 숫자는 주식분석가 의견수(몇명)
```
